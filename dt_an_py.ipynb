```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#filename = 'notebooks/data_an/Edu-Data.csv'
df = pd.read_csv('Edu-Data.csv')
```

```python
#visualizar o cabeçalho do dataset
df.head()
```

```python
#verifica as informações sobre o dataset e suas colunas
df.info()
```

```python
#Informações sobre o dataset, nesse método está presente a média e o desvio padrão
df.describe(include='all')
#df.describe()
```

```python
#Média, moda e mediana 

media_raised_hands = df['raisedhands'].mean()
print ('Média: ') 
print(media_raised_hands)

moda_raised_hands = df['raisedhands'].mode().iloc[0]
print('Moda: ')
print(moda_raised_hands)

mediana_raised_hands = df['raisedhands'].median()
print("Mediana: ")
print(mediana_raised_hands)
```

```python
#Achando dados escritos de forma errada, nesse caso, Venezuela estava escrito como "venzuela", e "Mom" como "Mum"
print(df[df['NationalITy']=='venzuela'])
#print(df[df['Relation']=='Mum'])
```

```python
##Inserir o dado anterior para demonstração
#df = df.replace('venezuela', 'venzuela', regex=True)
#df = df.replace('Mom', 'Mum', regex=True)
#df.head(64)
```

```python
#Substituindo esses dados
df = df.replace('venzuela', 'venezuela', regex=True)
df = df.replace('Mum', 'Mom', regex=True)
df.head(64)
```

```python
#Checando se há dados duplicados
df.duplicated()
```

```python
#Plotando um gráfico relacionando a frequência dos alunos com a satisfação dos pais com a escola

#Contamos quantas vezes aparece os dados "under-7" e "above-7"
a = df['StudentAbsenceDays'].mean()
print(a)

#Contamos a mesma coisa para satisfação do pais com a escola
b = df['ParentschoolSatisfaction'].value_counts()
print(b)
```

```python
#Usamos esses valores para calcular a porcentagem de alunos com faltas acima de 7 e de pais insatisfeitos com a escola

faltas = 39.79
insatisfeito = 39.16

#Definimos os rótulos para as barras
labels = ['Faltas dos Alunos', 'Insatisfação dos Pais']

#Definimos as porcentagens em uma lista
porcentagens = [faltas, insatisfeito]

#Definimos as posições das barras no eixo x
x = np.arange(len(labels))

#Criamos o gráfico de barras agrupadas
plt.bar(x, porcentagens)

#Adicionamos rótulos nos eixos e título
plt.ylabel('Porcentagem')
plt.title('Porcentagem de Faltas dos Alunos e Insatisfação dos Pais')

#Definimos o nome das métricas no eixo x
plt.xticks(x, labels)

#Colocamos o limite do eixo y em 100. Retirando esse limite é possível observar o gráfico mais de perto
plt.ylim(0, 100)

# Exibir o gráfico
plt.show()
```
